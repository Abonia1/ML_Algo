```python
# Regression Template
# Multiple Linear Regression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the datasets
datasets = pd.read_csv('Position_Salaries.csv')
X = datasets.iloc[:, 1:2].values
Y = datasets.iloc[:, 2].values
```

```python
"""# Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
"""
Y = Y.reshape(-1,1)

# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_Y = StandardScaler()
X = sc_X.fit_transform(X)
Y = sc_Y.fit_transform(Y)
```

```python
# Fitting the SVR model to the dataset

from sklearn.svm import SVR
regressorRdf = SVR(kernel = 'rbf')
regressorRdf.fit(X,Y)
```

```python
# Fitting the Linear SVR  model to the dataset
regressorLinear = SVR(kernel = 'linear')
regressorLinear.fit(X,Y)
# Predicting a new result with the Polynomial Regression 
Y_Pred = sc_Y.inverse_transform(regressorLinear.predict(sc_X.transform(np.array([6.5]).reshape(-1,1))))
```

```python
# Fitting the Polynomial SVR model to the dataset
regressorPolynomial = SVR(kernel = 'poly')
regressorPolynomial.fit(X,Y)
# Predicting a new result with the Polynomial Regression 
Y_Pred = sc_Y.inverse_transform(regressorPolynomial.predict(sc_X.transform(np.array([6.5]).reshape(-1,1))))
```

```python
# Visualising the Regression results

plt.scatter(X,Y, color = 'red')
plt.plot(X, regressorRdf.predict(X), color = 'blue')
plt.title('Polynomial Regression Results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Regression results

plt.scatter(X,Y, color = 'red')
plt.plot(X, regressorLinear.predict(X), color = 'blue')
plt.title('Linear Regression Results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Regression results

plt.scatter(X,Y, color = 'red')
plt.plot(X, regressorPolynomial.predict(X), color = 'blue')
plt.title('RDF Regression Results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
